```python
import pandas as pd

TRAINING_DATA_PATH = '../data_processed/training_data_NEW.csv'
TEST_DATA_PATH = '../data_processed/test_data_NEW.csv'

def load_data(data_path):
    return pd.read_csv(data_path)

music_train = load_data(TRAINING_DATA_PATH)
music_test = load_data(TEST_DATA_PATH)
music_train.head()
```
